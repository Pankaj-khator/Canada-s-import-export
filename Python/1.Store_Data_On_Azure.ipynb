# Uploading all csv files from local storage to Azure Storage


```python
import pandas
from azure.storage.blob import BlobServiceClient
import os
from dotenv import load_dotenv
```

```python
os.getcwd()
```

## Accessing credentials through .env

```python
load_dotenv()
account_url = os.environ.get("ACCOUNT_URL")
SAS_token = os.environ.get("SAS_token")
container = os.environ.get("container")
```

```python
blobService = BlobServiceClient(
    account_url=account_url,
    credential=SAS_token,
    #max_chunk_get_size = 10000
)
```

```python
# Looping all the imports files to store on the azure
import pandas as pd
import pyarrow
from io import BytesIO
for i in range(2002, 2023):
    df = pd.read_csv(
        f'/Users/pankajkumar/Library/CloudStorage/OneDrive-NC/Business Analytics/1st Term/2nd Term/Capstone Project/Restart/Data/Import/N{i}_I.csv'
    )

    blob_name = f"Import_{i}.parquet"

    blob_client = blobService.get_blob_client(
        container=container,
        blob=blob_name
    )

    parquet_buffer = BytesIO()
    df.to_parquet(parquet_buffer, engine="pyarrow", index=False)
    parquet_buffer.seek(0)

    blob_client.upload_blob(parquet_buffer, overwrite=True)

    print(blob_name, "Uploaded")
```
