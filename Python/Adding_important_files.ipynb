```python
import pandas as pd
from io import BytesIO
import os
import pyarrow
import azure.storage.blob 
from dotenv import load_dotenv
load_dotenv()
```

```python

blobservice = azure.storage.blob.BlobServiceClient(
    account_url=os.environ.get('ACCOUNT_URL'),
    credential=os.environ.get('SAS_token')
)
```

```python
country_and_continent = pd.read_csv('/Users/pankajkumar/Library/CloudStorage/OneDrive-NC/Business Analytics/1st Term/2nd Term/Capstone Project/Restart/Data/Usefull/Country.csv')
```

```python
HS6_codes = pd.read_excel('/Users/pankajkumar/Library/CloudStorage/OneDrive-NC/Business Analytics/1st Term/2nd Term/Capstone Project/Restart/Data/Usefull/HS6.xlsx')
```

```python
HS6_codes = HS6_codes.drop(columns=['sub_cat'])
```

```python

parquet_buffer = BytesIO()
country_and_continent.to_parquet(parquet_buffer, engine='pyarrow')
parquet_buffer.seek(0)

blobclient = blobservice.get_blob_client(
    container=os.environ.get('container'),
    blob = 'country_and_continent.parquet'
)

blobclient.upload_blob(parquet_buffer, overwrite=True)
print('country_and_continent Uploaded')
```

```python

parquet_buffer = BytesIO()
HS6_codes.to_parquet(parquet_buffer, engine='pyarrow')
parquet_buffer.seek(0)

blobclient = blobservice.get_blob_client(
    container=os.environ.get('container'),
    blob = 'HS6_codes.parquet'

)

blobclient.upload_blob(parquet_buffer, overwrite=True)
print('HS6_codes Uploaded')
```
